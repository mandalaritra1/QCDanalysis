## Example of making ratio plots

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Generate some random data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Compute the ratio of y1 and y2
ratio = y1 / y2

# Create a figure with a gridspec of two rows with different heights
fig = plt.figure()
gs = gridspec.GridSpec(nrows=2, ncols=1, height_ratios=[4, 1], hspace = 0.15)

# Plot y1 and y2 on the first subplot using gs[0]
ax1 = plt.subplot(gs[0])
ax1.plot(x, y1, label='y1', color = 'r')
ax1.plot(x, y2, label='y2', color = 'b')
ax1.set_ylabel('Quantities')
ax1.set_xscale('log')
ax1.set_xticklabels('')


ax1.tick_params(axis = 'x', top = True, right =True, direction = 'in')
ax1.tick_params(axis = 'y', right = True, direction = 'in')

ax1.minorticks_on()
ax1.tick_params(axis = 'x', which = 'minor', direction = 'in', top = True)
ax1.tick_params(axis = 'y', which = 'minor', direction = 'in', right = True)

ax1.xaxis.set_label_coords(1.05, -0.025)
ax1.yaxis.set_label_coords(-0.1, 0.85)

ax1.legend()

# Plot the ratio on the second subplot using gs[1]
ax2 = plt.subplot(gs[1])
ax2.plot(x, ratio, label='y1/y2', color = 'black')
ax2.set_xlabel('x')
ax2.set_ylabel('Ratio')
ax2.legend()
ax2.tick_params(axis = 'x', top = True, right =True, direction = 'in')
ax2.tick_params(axis = 'y', right = True, direction = 'in')
ax2.xaxis.set_label_coords(0.95, -0.3)

ax2.minorticks_on()
ax2.tick_params(axis = 'x', which = 'minor', direction = 'in', top = True)
ax2.tick_params(axis = 'y', which = 'minor', direction = 'in', right = True)



# Show the plot
plt.show()
```

```python

def adjust_plot(ax1,title = None, xlabel = None, ylabel = None, xscale = None, yscale = None, minor_ticks = True, xlabel_coord = 0.95, ylabel_coord = 0.9 ):
    if xlabel != None:
        ax1.set_xlabel(xlabel)
    if ylabel != None:
        ax1.set_ylabel(ylabel)
    if minor_ticks == True:
        ax1.minorticks_on()
    if xscale != None:
        ax1.set_xscale(xscale)
    if yscale != None:
        ax1.set_yscale(xscale)
        
    if title != None:
        ax1.set_title(title)
    ax1.tick_params(axis = 'x', top = True, direction = 'in', length = 5)
    ax1.tick_params(axis = 'y', right = True, direction = 'in', length = 5)
    ax1.tick_params(axis = 'x', which = 'minor', direction = 'in', top = True, length = 2.5)
    ax1.tick_params(axis = 'y', which = 'minor', direction = 'in', right = True, length = 2.5)
    ax1.xaxis.set_label_coords(xlabel_coord, -0.06)
    ax1.yaxis.set_label_coords(-0.1, ylabel_coord)
    ax1.legend()
    
    
```

```python
ax = plt.subplot()
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
plt.plot(x,y1, label = 'Test', color = 'r')

adjust_plot(ax, xlabel = r"Tumi", ylabel = r"Ami")
```
